```python
import numpy as np
import gzip

# load training images
with gzip.open(r'E:\AIO\Multilayer_Perception\Data_Fashion_Mnist\train-images-idx3-ubyte.gz', 'rb') as f:
    X_train = np.frombuffer(f.read(), np.uint8, offset=16).reshape(-1, 28*28)

# load testing images
with gzip.open(r'E:\AIO\Multilayer_Perception\Data_Fashion_Mnist\t10k-images-idx3-ubyte.gz', 'rb') as f:
    X_test = np.frombuffer(f.read(), np.uint8, offset=16).reshape(-1, 28*28)

# load training labels
with gzip.open(r'E:\AIO\Multilayer_Perception\Data_Fashion_Mnist\train-labels-idx1-ubyte.gz', 'rb') as f:
    y_train = np.frombuffer(f.read(), np.uint8, offset=8)
    
# load testing labels
with gzip.open(r'E:\AIO\Multilayer_Perception\Data_Fashion_Mnist\train-labels-idx1-ubyte.gz', 'rb') as f:
    y_test = np.frombuffer(f.read(), np.uint8, offset=8)
    

# check data
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

# Tạo danh sách 100 phần tử ngẫu nhiên từ X_train có 60000 phần tử
indices = list(np.random.randint(X_train.shape[0],size=144))

fig =plt.figure(figsize=(9,9))
columns = 12
rows = 12
for i in range(1, columns*rows +1):
    img = X_train[indices[i-1]].reshape(28,28)
    fig.add_subplot(rows, columns, i)
    plt.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```
